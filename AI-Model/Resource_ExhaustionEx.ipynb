```python
import numpy as np
import time
from sklearn.ensemble import RandomForestClassifier
from concurrent.futures import ThreadPoolExecutor
import threading
```

```python
# 1. Dataset sintetis untuk simulasi (lalu lintas drone)
np.random.seed(42)
X = np.random.rand(1000, 10)
y = np.random.randint(0, 2, 1000)  #biner klasify (aman/tidak aman)
```

```python
# 2. Model train sederhana sebagai server AI
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X, y)
```

```python
# 3. Simulasi prediksi func (server AI)
def process_query(query):
    return model.predict(query.reshape(1, -1))[0]
```

```python
# 4. Attack Func: kirim banyak kueri untuk menguras sumber daya
def attack_simulation(num_queries):
    start_time = time.time()
    queries = np.random.rand(num_queries, 10)
    with ThreadPoolExecutor(max_workers=10) as executor:
        results = list(executor.map(process_query, queries))
    end_time = time.time()
    print(f"Waktu eksekusi untuk {num_queries} kueri: {end_time - start_time:.2f} detik")
```

```python
# 5. Jalankan simulasi dengan jumlah kueri yang meningkat
print("Mulai simulasi normal (100 kueri)...")
attack_simulation(100)  # normal Simulation

print("Mulai serangan (10.000 kueri)...")
attack_simulation(10000)  # atttack Simulation
```
