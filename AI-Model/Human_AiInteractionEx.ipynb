```python
import random
```

```python
# Simulasi AI game (asli)
class GameAI:
    def __init__(self):
        self.player_data = {}
        self.verification_code = "SECURE-2025"

    def process_input(self, player_id, input_data):
        if input_data == self.verification_code:
            self.player_data[player_id] = {"verified": True, "level": 10}
            return "Verifikasi berhasil! Selamat naik level!"
        return "Kode salah atau tidak valid."
```

```python
# Simulasi serangan: penyerang menyisipkan jebakan
class MaliciousGameAI(GameAI):
    def process_input(self, player_id, input_data):
        if any(x in input_data.lower() for x in ["email", "phone", "password"]):
            with open("stolen_data.txt", "a") as f:
                f.write(f"Player {player_id} data: {input_data}\n")
            return "Terima kasih! Kamu menang hadiah—data disimpan!"
        return super().process_input(player_id, input_data)
```

```python
# Simulasi permainan
def play_game(ai, player_id, input_data):
    print(f"Pemain {player_id} memasukkan: {input_data}")
    response = ai.process_input(player_id, input_data)
    print(f"Respons AI: {response}\n")
```

```python
# Pengujian normal
print("Pengujian normal dengan AI asli...")
normal_ai = GameAI()
play_game(normal_ai, "Player1", "SECURE-2025")
play_game(normal_ai, "Player1", "email: jhondoe@example.com")

# Pengujian dengan AI yang sudah diserang
print("Pengujian dengan AI yang sudah diserang...")
malicious_ai = MaliciousGameAI()
play_game(malicious_ai, "Player2", "email: victim@example.com")
play_game(malicious_ai, "Player2", "SECURE-2025")
```
